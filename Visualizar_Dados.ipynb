```python
import sqlite3
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def plotar_grafo_banco_dados(banco_dados, tabela, coluna_origem, coluna_destino, caso_analise):

    conexao = sqlite3.connect(banco_dados)
    cursor = conexao.cursor()

    query = f"SELECT {coluna_origem}, {coluna_destino} FROM {tabela} WHERE Caso = '{caso_analise}'"
    cursor.execute(query)
    relacionamentos = cursor.fetchall()
    conexao.close()

    G = nx.DiGraph()
    G.add_edges_from(relacionamentos)
    pos = nx.spring_layout(G)

    plt.figure(figsize=(8, 6))
    nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000,
            font_size=10, edge_color='gray', arrows=True, arrowsize=20)
    
    plt.tight_layout()
    
    plt.savefig('Grafo - ' + caso_analise + '.png', format='png', dpi=300, bbox_inches='tight')
    plt.show()



def plotar_grafo_direcionado(relacionamentos, arquivo="Grafo Direcionado.png"):

    G = nx.DiGraph()
    G.add_edges_from(relacionamentos)
  
    pos = nx.spring_layout(G)
    
    plt.figure(figsize=(8, 6))
    nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000,
            font_size=10, edge_color='gray', arrows=True, arrowsize=20)
    
    plt.tight_layout()
    
    plt.savefig(arquivo, format='png', dpi=300, bbox_inches='tight')
    plt.show()
```

```python
plotar_grafo_banco_dados(r'C:\Stage\Projects\TCC_Univesp\TCC_Univesp.db', 'Estudo_Caso', 'Origem', 'Destino', 'WEVERTON ROCHA MARQUES DE SOUSA x EVILSON PINTO DE ALMEIDA SOBRINHO')
```

```python
relacionamentos = [("Evilson Sobrinho", "Doao"),
                   ("Doao", "Weverton Rocha"),
                   ("Weverton Rocha", "Codevasf"),
                   ("Codevasf", "Contratos"),
                   ("Contratos", "Viacom"),
                   ("Viacom", "Evilson Sobrinho"),]

plotar_grafo_direcionado(relacionamentos)
```
